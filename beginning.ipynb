```python
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()

client = Anthropic()  # tự đọc ANTHROPIC_API_KEY từ .env

def chat(user_message):
    message = client.messages.create(
        model="claude-haiku-4-5",   # model rẻ nhất
        max_tokens=100,              # giới hạn độ dài phản hồi
        system="Trả lời ngắn gọn, tối đa 3 câu.",
        messages=[
            {"role": "user", "content": user_message}
        ]
    )

    print("Claude:", message.content[0].text)
    print("---")
    print(f"Tokens dùng: {message.usage.input_tokens} input / {message.usage.output_tokens} output")

# Test thử
chat("Xin chào! Bạn là ai?")
```
